# What's a clique?

A subset of nodes of an undirected graph where any pair of nodes is linked. Maybe another way to see this is... a "comunity" where all nodes are connected with all nodes: subgraph complete.

In social networks: groups of people all of whom known each other.

Maximal clique: the clique that cannot be extended adding connected (adjacent) nodes.
Maximum clique: the clique with the largest number of nodes.
Intersection number of the graph: the smallest number of cliques covering all graph edges.
Clique cover number of a graph: the smallest number of cliques covering all graph nodes.
